__1.__ Euler method to solve

$\frac{dy}{dt}=y-t$

at time $t=1$ given that $y(0)=\frac{1}{2}$ in 3 steps. 

Redo this estimate with a fully implicit scheme. 

Note that the true solution is $y=0.640859$

```python
ti=0
tf=1
nsteps=3
yi=0.5

t=np.linspace(ti,tf,nsteps+1)

dt=t[1]-t[0]

y=np.zeros(nsteps+1)

y[0]=yi

for i in range(nsteps):
    y[i+1]=y[i]+dt*(y[i]-t[i])
    
print(f'The value calculated of y at t={t[-1]} is y={y[-1]}')
    
figure=plt.figure(figsize=(12,8))
plt.plot(t,y)
plt.xlabel('t')
plt.ylabel('y')
```

```python
yimp=np.zeros(nsteps+1)

yimp[0]=yi

for i in range(nsteps):
    yimp[i+1]=(yimp[i]-dt*(t[i+1]))/(1-dt)
    
print(f'The value calculated of y (via the implicit scheme) at t={t[-1]} is y={y[-1]}')
    
figure=plt.figure(figsize=(12,8))
plt.plot(t,yimp)
plt.xlabel('t')
plt.ylabel('y')
```
